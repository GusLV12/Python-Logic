# Valores Atipicos

Estos valores pueden dañar ña preision de nuestro codigo. Debido a que son valores especiales de los caules pueden dañar la media, mediana y moda.

```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
```

```python
#datos atipicos
data = np.array([10, 20, 30, 12, 42, 53, 63, 30, 72, 18, 19, 10, 59])
```

```python
# Graficar los datos
plt.boxplot(data, vert=True)
plt.title("Boxplot de datos con outliers")
plt.xlabel("Valores")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
iqr = Q3 - Q1
```

```python
# Filtrar los datos atípicosn (outliers)

limite_inferior = Q1 - 1.5 * iqr
limite_superior = Q3 + 1.5 * iqr
data_filter = data[(data >= limite_inferior) & (data <= limite_superior)]

plt.boxplot(data_filter,vert=True)
plt.title("Boxplot de datos filtrados")
plt.xlabel("Valores filtrados")
print(limite_inferior)
print(limite_superior)
plt.show()
```
