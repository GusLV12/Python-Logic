# Repaso (creo)

### Media
La media es el promedio de un conjunto de datos y representa el valor central de una distribuciÃ³n. Se calcula sumando todos los valores y dividiÃ©ndolos por el nÃºmero total de elementos.

**FÃ³rmula**:  
$$
\text{Media} = \frac{1}{N} \sum_{i=1}^{N} X_i
$$  
donde \( N \) es el nÃºmero de datos y \( X_i \) representa cada valor.

### DispersiÃ³n
La dispersiÃ³n mide cuÃ¡nto varÃ­an los datos respecto a la media, indicando la extensiÃ³n o variabilidad de los valores. Algunas medidas comunes son la varianza y el rango intercuartÃ­lico.

(No tiene una fÃ³rmula especÃ­fica, ya que es un concepto general que agrupa varias medidas).

### Varianza
La varianza es el promedio de los cuadrados de las desviaciones respecto a la media. Cuanto mayor es la varianza, mayor es la dispersiÃ³n de los datos.

**FÃ³rmula**:  
$$
\text{Varianza} = \frac{1}{N} \sum_{i=1}^{N} (X_i - \text{Media})^2
$$  

### DesviaciÃ³n EstÃ¡ndar
La desviaciÃ³n estÃ¡ndar es la raÃ­z cuadrada de la varianza y mide cuÃ¡nto se alejan en promedio los datos respecto a la media. Es Ãºtil para entender la dispersiÃ³n en unidades originales de los datos.

**FÃ³rmula**:  
$$
\text{DesviaciÃ³n EstÃ¡ndar} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (X_i - \text{Media})^2}
$$


## Evento de Probabilidad
Un evento de probabilidad es un conjunto de posibles resultados de un experimento aleatorio. Especifica las circunstancias bajo las cuales ocurren ciertos resultados, como obtener "cara" al lanzar una moneda. Los eventos se representan con variables aleatorias, que asignan valores numÃ©ricos a los posibles resultados de un experimento.

## Variables Aleatorias
Una variable aleatoria es una funciÃ³n que asigna un valor numÃ©rico a cada resultado de un experimento. Las variables aleatorias permiten modelar y cuantificar la ocurrencia de eventos, facilitando el cÃ¡lculo de probabilidades en distintos escenarios.

## Ensayo de Bernoulli
Un ensayo de Bernoulli es un experimento aleatorio con solo dos resultados posibles: Ã©xito o fracaso (generalmente representados como 1 y 0, respectivamente). Cada ensayo tiene una probabilidad fija de Ã©xito \( p \), y los resultados son independientes de otros ensayos. Este tipo de ensayo es la base de la distribuciÃ³n binomial.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df'
# Select only the numeric columns
numeric_cols = chocolate.select_dtypes(include='number').columns

# Plot the distribution of each numeric column
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(chocolate[col], kde=True)  
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Selecting non-numeric columns
non_numeric_cols = chocolate.select_dtypes(exclude=['number']).columns

# Plotting distribution of each non-numeric column
for col in non_numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=chocolate[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
for col in non_numeric_cols:
    plt.figure(figsize=(8, 4))
    
    # Get the top 10 most frequent values
    top_values = chocolate[col].value_counts().nlargest(10).index
    top_data = chocolate[chocolate[col].isin(top_values)]
    
    # Plot
    sns.countplot(x=top_data[col], order=top_values)
    plt.title(f'Top 10 Most Repeated Values in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
for col in non_numeric_cols:
    plt.figure(figsize=(8, 4))
    
    # Get the top 10 most frequent values
    top_values = chocolate[col].value_counts().nsmallest(10).index
    top_data = chocolate[chocolate[col].isin(top_values)]
    
    # Plot
    sns.countplot(x=top_data[col], order=top_values)
    plt.title(f'Top 10 Less Repeated Values in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()
```

# Distribuciones de probabilidad

| **DistribuciÃ³n**                 | **DescripciÃ³n**                                                                 | **Propiedades MatemÃ¡ticas**                                                                                                    | **Aplicaciones en ML**                                                                                      |
|----------------------------------|---------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------|
| **DistribuciÃ³n Normal** (Gaussiana)  | DistribuciÃ³n continua caracterizada por una curva en forma de campana, simÃ©trica alrededor de la media.               | - Definida por la media (Î¼) y la desviaciÃ³n estÃ¡ndar (Ï).<br>- El 68% de los datos cae dentro de 1Ï, el 95% dentro de 2Ï, y el 99.7% dentro de 3Ï. | - Escalado de caracterÃ­sticas (normalizaciÃ³n, estandarizaciÃ³n).<br>- Clasificador de Naive Bayes Gaussiano.<br>- Pruebas de suposiciÃ³n (por ejemplo, pruebas t). |
| **DistribuciÃ³n Binomial**             | DistribuciÃ³n discreta que representa el nÃºmero de Ã©xitos en un nÃºmero fijo de ensayos independientes.    | - Definida por el nÃºmero de ensayos (n) y la probabilidad de Ã©xito (p).<br>- FunciÃ³n de masa de probabilidad: <br> \(P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}\) | - ClasificaciÃ³n binaria (por ejemplo, regresiÃ³n logÃ­stica).<br>- Pruebas A/B para comparar versiones.<br>- EvaluaciÃ³n de mÃ©tricas de modelos (precisiÃ³n, recall). |
| **DistribuciÃ³n de Poisson**           | DistribuciÃ³n discreta para el nÃºmero de eventos en un intervalo fijo, dado un promedio conocido (Î»).    | - Definida por la tasa promedio (Î»).<br>- FunciÃ³n de masa de probabilidad: <br> \(P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}\) | - Modelado de datos de conteo (por ejemplo, clics, ventas).<br>- DetecciÃ³n de anomalÃ­as basada en conteos de eventos.<br>- RegresiÃ³n de Poisson para predicciÃ³n de conteos. |


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Parameters for the Normal Distribution
mu, sigma = 0, 1  # Mean and standard deviation

# Generate data
data = np.random.normal(mu, sigma, 1000)

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Probability Density Function (PDF)
sns.histplot(data, bins=30, kde=True, stat='density', ax=axes[0])
axes[0].set_title('Normal Distribution - PDF')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')

# 2. Cumulative Distribution Function (CDF)
sns.ecdfplot(data, ax=axes[1])
axes[1].set_title('Normal Distribution - CDF')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Cumulative Probability')

# 3. Histogram with fitted distribution
sns.histplot(data, bins=30, stat='density', ax=axes[2])
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
axes[2].plot(x, norm.pdf(x, mu, sigma), color='red', lw=2)
axes[2].set_title('Normal Distribution - Histogram with Fitted PDF')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

# Parameters for the Binomial Distribution
n, p = 10, 0.5  # Number of trials and probability of success

# Generate data
data = binom.rvs(n, p, size=1000)

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Probability Mass Function (PMF)
x = np.arange(0, n+1)
axes[0].bar(x, binom.pmf(x, n, p), color='blue', alpha=0.7)
axes[0].set_title('Binomial Distribution - PMF')
axes[0].set_xlabel('Number of Successes')
axes[0].set_ylabel('Probability')

# 2. Cumulative Distribution Function (CDF)
axes[1].bar(x, binom.cdf(x, n, p), color='orange', alpha=0.7)
axes[1].set_title('Binomial Distribution - CDF')
axes[1].set_xlabel('Number of Successes')
axes[1].set_ylabel('Cumulative Probability')

# 3. Histogram with fitted PMF
sns.histplot(data, bins=n, stat='density', ax=axes[2], kde=False, color='lightgray')
axes[2].bar(x, binom.pmf(x, n, p), color='blue', alpha=0.7)
axes[2].set_title('Binomial Distribution - Histogram with Fitted PMF')
axes[2].set_xlabel('Number of Successes')
axes[2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Parameters for the Poisson Distribution
lambda_ = 3  # Average rate of occurrence

# Generate data
data = poisson.rvs(lambda_, size=1000)

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. Probability Mass Function (PMF)
x = np.arange(0, 15)
axes[0].bar(x, poisson.pmf(x, lambda_), color='green', alpha=0.7)
axes[0].set_title('Poisson Distribution - PMF')
axes[0].set_xlabel('Number of Events')
axes[0].set_ylabel('Probability')

# 2. Cumulative Distribution Function (CDF)
axes[1].bar(x, poisson.cdf(x, lambda_), color='purple', alpha=0.7)
axes[1].set_title('Poisson Distribution - CDF')
axes[1].set_xlabel('Number of Events')
axes[1].set_ylabel('Cumulative Probability')

# 3. Histogram with fitted PMF
sns.histplot(data, bins=30, stat='density', ax=axes[2], kde=False, color='lightgray')
axes[2].bar(x, poisson.pmf(x, lambda_), color='green', alpha=0.7)
axes[2].set_title('Poisson Distribution - Histogram with Fitted PMF')
axes[2].set_xlabel('Number of Events')
axes[2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
